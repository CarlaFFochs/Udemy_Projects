### Make a plot from dictionary

```python
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title = "Dan's Nut Butter Store",
       ylabel = "Price ($)");
```

### Two options for subplots

#### Subplot option 1 (recommended)

```python
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows = 2,
                                            ncols= 2,
                                            figsize = (10, 5))
# plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

#### Subplot option 2

```python
fig, ax = plt.subplots(nrows= 2,
                   ncols=2,
                   figsize = (10,5))
# Plot to each different index
ax[0,0].plot(x, x/2) #row 0, column 0
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one should you use ? (pyplot vs matplotlib OO method)?

* When plotting something quickly, okay to use pyplot method
* When plottinf more advanced, use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50.shape[0]
```

```python
over_50.plot(kind="scatter",
             x= "age", 
             y= "chol",
             c= "target");
```

```python
# not much clear with the colors... let's customized better in order to see clearly.
```

```python
## OO method mized with pyplt
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind= "scatter",
             x= "age",
             y = "chol",
             c= "target",
             ax = ax); # plt the data on the axis ax
# ax.set_xlim([45,100])  # to change the range of the axis         
```

```python
# Let's customized
## OO method from scratch
fig, ax = plt.subplots(figsize = (10,6))

# Plot the data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"]); # to color the target

# Customize the plot
ax.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title = "Target")
# grab the values from "scatter" and unpack the legend elements (is going to query the "c" parameter and is going to see the target column and they us going to create the legend elements)

# Add an horizontal line across the axis
ax.axhline(over_50["chol"].mean(), 
           linestyle= "--");
```

```python
over_50.head()
```

```python
# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows= 2,
                               ncols = 1,
                               figsize = (10,10),
                               sharex=True)
# Add data to to ax0
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"])
# Customize ax0
ax0.set(title= "Heart Disease and Cholesterol Levels", #when we include "sharex = True", we have to remove the "xlabel"
        ylabel = "Cholesterol")
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title = "Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle= "--");

# Add data to ax1
scatter = ax1.scatter(x= over_50["age"],
                      y= over_50["thalach"],
                      c= over_50["target"])
# Customize ax11
ax1.set(title = "Heart Disease and Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate");
# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title= "Target")
# Add a meanline
ax1.axhline(y= over_50["thalach"].mean(),
            linestyle = "--");
# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight="bold");
```

```python
# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows= 2,
                               ncols = 1,
                               figsize = (10,10),
                               sharex=True)
# Add data to to ax0
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"],
                      cmap = "winter")
# Customize ax0
ax0.set(title= "Heart Disease and Cholesterol Levels", #when we include "sharex = True", we have to remove the "xlabel"
        ylabel = "Cholesterol")
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title = "Target")

# Change the x axis limits
ax0.set_xlim([50,80])

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle= "--");

# Add data to ax1
scatter = ax1.scatter(x= over_50["age"],
                      y= over_50["thalach"],
                      c= over_50["target"],
                      cmap = "winter")
# Customize ax11
ax1.set(title = "Heart Disease and Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate");
# Change ax1 x axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title= "Target")
# Add a meanline
ax1.axhline(y= over_50["thalach"].mean(),
            linestyle = "--");
# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight="bold");
```
